```python
import pandas as pd
import numpy as np
def conv(val):
    if val == np.nan:
        return 0 # or whatever else you want to represent your NaN with
    return val
cars = pd.read_csv('C:/Users/Anima/OneDrive/Рабочий стол/R_PythonCourseKNU/Labs/Python_Lab2/Auto_Owners.csv',converters={"OWN_WEIGHT" : conv, "TOTAL_WEIGHT" : conv} , delimiter = ';',error_bad_lines=False)
cars.head()
```

### Питання 1
Знайти автомобіль з найбільшою зареєстрованою місткості

```python
def One():
    return cars[cars['CAPACITY'] == cars['CAPACITY'].max()]
One()
```

### Питання 2
Знайти кількість людей що купи найпопулярніший клас автомобіля

```python
def answer_two():
    x = cars.groupby("BODY").count()
    return x["PERSON"].max()
answer_two()
```

### Питання 3
Знайти топ 5 брендів за популярністю та кількість їх покупок

```python
def answer_three():
    return cars['BRAND'].value_counts()[:5].sort_values(ascending=False)
answer_three()
```

### Питання 4
Знайти відчизняні автомобілі які були вироблені після 2015 року

```python
def answer_four():
    return cars[(cars['BRAND'] == 'ЗАЗ') & (cars['MAKE_YEAR']>2015)]
answer_four()
```

### Питання 5
Знайти відношення операцій з відчизняними автомобілями до операцій з інномарками. 

```python
def answer_five():
    local = cars[cars['BRAND'] == 'ЗАЗ'].count()
    foreign = ~cars[cars['BRAND'] == 'ЗАЗ'].count()
    res = local/foreign
    return res['PERSON']
answer_five()
```
